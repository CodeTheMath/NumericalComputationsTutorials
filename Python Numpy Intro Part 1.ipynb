```python
# -- imports
import numpy as np
from matplotlib import pyplot as plt
```

```python
# -- basic datatypes in Python

# -- dict
my_dict = {}
my_dict['key1'] = 1
my_dict['key2'] = 'string'
print(f'this is my dict {my_dict}')
key = 'key2'
print(f'this is my dict {my_dict[key]}')
keys = my_dict.keys()
print('==========================')


# -- list
li = ['aa', 1, {'key': 'dog'}]
li[-1]['key']

# -- set
print('=================')
my_set = {1,2,3,4,5,1,6}
print(my_set)
li = [1,1,1,2,3,4]
my_set = set(li)
print(my_set)

# -- tuple
tu = (1,2,3)
# tu[0] = 10
```

```python
# -- numpy arrays and functions

# -- zeros
zeros_ar = np.zeros((2,2))
print(f'my zeros {zeros_ar.shape}')

# -- ones
my_ones = np.ones((2,2))
# my_ones

# -- eye
my_eye = np.eye(2)

# -- creating array by hand
my_ar = np.array([[1,2],[3,4]])
my_ar

# -- creating array by concatenating building blocks
next_matrix = np.vstack([np.hstack([zeros_ar, my_ones]), np.hstack([my_eye, my_ar])])
next_matrix
```

```python
# -- make a simple 2D plot with matplotlib
x = np.arange(0, 10, 1E-1)

# -- compute sine()
y = np.sin(x)

plt.plot(x, y)
plt.xlabel('time')
plt.ylabel('values')
plt.title('simple plot')
plt.grid()
```

```python
# -- use an anonymous function to make simple computations

f = lambda arg: np.exp(arg) * np.sin(arg)
y = f(x)

plt.plot(x, y)
plt.xlabel('time')
plt.ylabel('values')
plt.title('simple plot')
```
